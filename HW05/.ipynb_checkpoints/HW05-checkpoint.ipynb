## Problem 1 
### Linear congruential random number generator:

#### 'gen_ random' function generates random number by using the following paramaters:
#### $a=5$ , $c=0 $ and $x0 =1$
#### 'gen_ random_ibm' function generates random number by using IBM parameters 

```python
%matplotlib inline
import numpy as np
from matplotlib import pyplot
```

```python
# Random number genrator  with certain parameters 


def gen_random (x0,N):
    a = 5 
    M = 11
    x = np.zeros(N)
    x[-1] = x0
    for i in range (0,N):
        x[i] = (a * x[i-1] ) % M
    for i in range (0,N):
        x[i]=float(x[i]) / float (M)
    return x

# Random number genrator  with IBM parameters 

def gen_random_ibm (x0,N): 
    a = 65539 
    M =  2**31
    x = np.zeros(N)
    x[-1] = x0
    for i in range (0,N):
        x[i] = (a * x[i-1] ) % M
    for i in range (0,N):
        x[i]=float(x[i]) / float (M)
    return x


```

### $a=5$ , $c=0 $ 

```python
# Using gen_random

ngroups = 16

I = np.zeros(ngroups)
N = np.zeros(ngroups)
E = np.zeros(ngroups)

n0 = 100
x0=1 # seed 
y=[x0]

for i in range(ngroups):

    N[i] = n0
    x = gen_random(y[-1],n0)
    #print (x)
    y = gen_random(x[-1],n0)
    #print(y)
    I[i] = 0.
    Nin = 0
    for j in range(n0):
        if(y[j] < np.sqrt(1-x[j]**2)):
            Nin += 1
            
    I[i] = 4.*float(Nin)/float(n0)
    E[i] = abs(I[i]-np.pi)
    print (n0,Nin,I[i],E[i])
    n0 *= 2
    
            
pyplot.plot(N,E,ls='-',c='red',lw=3,label='E(N)');
pyplot.plot(N,1/np.sqrt(N),ls='-',c='blue',lw=3, label='1/sqrt(N)');
pyplot.xscale('log')
pyplot.yscale('log')
pyplot.legend()
pyplot.title('a=5 , M=11')
```

### IBM parameters 

```python
ngroups = 16

I = np.zeros(ngroups)
N = np.zeros(ngroups)
E = np.zeros(ngroups)

n0 = 100
x0=1 # seed 
y=[x0]
for i in range(ngroups):

    N[i] = n0
    x = gen_random_ibm(y[-1],n0)
    y = gen_random_ibm(x[-1],n0)
    I[i] = 0.
    Nin = 0
    for j in range(n0):
        if(y[j] < np.sqrt(1-x[j]**2)):
            Nin += 1
            
    I[i] = 4.*float(Nin)/float(n0)
    E[i] = abs(I[i]-np.pi)
    print (n0,Nin,I[i],E[i])
    n0 *= 2
    
            
pyplot.plot(N,E,ls='-',c='red',lw=3,label='E(N)');
pyplot.plot(N,1/np.sqrt(N),ls='-',c='blue',lw=3, label='1/sqrt(N)');
pyplot.xscale('log')
pyplot.yscale('log')
pyplot.legend()
pyplot.title('IBM parameters')
```

## Problem 2 


$\int _0^1{x^2dx} = \frac{1}{3}= 0.3333$
#### Simple MC integration :

```python
#Boundries
a=0
b=1
h=1

N=1000000 # total number of points 
Nin=0 # number of acceptance
x=np.random.random(N)
y=np.random.random(N)

for i in range (0,N):
    if (y[i] < x[i]**2):
        Nin += 1
        
I = (b-a) * h * Nin / N 

print ('*****************\n I =',I,'\n*****************')

```

### Importance sampling 
 $p(x)=2x$ 
 
 $\int _0^1{p(x) dx} =1 $
 
 $y(x)= \sqrt x$
 
 $ I = \frac{1}{N} \sum_{i=1}^N \frac{y_i^2}{2y_i} =  \frac{1}{N} \sum_{i=1}^N \frac{y_i}{2}$
 

```python
N=1000000

x=np.random.random(N)

y=np.sqrt(x)

I =0 

for i in range(0,N):
    I += y[i]/2

I=I/N

print ('*************************\n I =',I,'\n*************************')
```

## Problem 3


$\int _0^1{\sqrt x dx} = \frac{2}{3}= 0.666667$
#### Simple MC integration :

```python
#Boundries
a=0
b=1
h=1

N=1000000 # total number of points 
Nin=0 # number of acceptance
x=np.random.random(N)
y=np.random.random(N)

for i in range (0,N):
    if (y[i] < np.sqrt(x[i])):
        Nin += 1
        
I = (b-a) * h * Nin / N 

print ('*****************\n I =',I,'\n*****************')

```

### Importance sampling 
 $p(x)=2x$ 
 
 $\int _0^1{p(x) dx} =1 $
 
 $y(x)= \sqrt x$
 
 $ I = \frac{1}{N} \sum_{i=1}^N \frac{y_i^2}{2y_i} =  \frac{1}{N} \sum_{i=1}^N \frac{y_i}{2}$
 

```python
N=100000
a=1
A = (1 - np.exp(-a))  # Normalization constant inverse  
x=np.random.random(N)

y = -np.log(x) / a

I =0 

for i in range(0,N):
    I +=   A * np.sqrt(y[i])/ (a * np.exp(-a*y[i]))

I=I/N

print ('*************************\n I =',I,'\n*************************')
```
